```python
#Part 3:

#1) Write (Python) code to generate annual summary statistics per team â€“ regular season only

#First, we are creating a column known as 'Team and Year' that will help us combine the two tables and group into team stats
stats_season_list = [str(x) for x in stats1.column('Season')] 
stats_team_list = [x for x in stats1.column('Team')]
stats_team_year = make_array() 
for i in range(len(stats_season_list)): #This step takes a bit long to run so please be patient
    temp1 = stats_team_list[i] + ' '+ stats_season_list[i]
    stats_team_year = np.append(stats_team_year, temp1)

#Now we do the same for 'results'
results_year_list = [str(x) for x in results.column('Year')]
results_team_list = [x for x in results.column('Team')]
results_team_year = make_array()
for i in range(len(results_year_list)):
    temp2 = results_team_list[i] + ' ' + results_year_list[i]
    results_team_year = np.append(results_team_year, temp2)
```

```python
#3) Write code to help determine which players contributed the most to each team on a position-wise basis - regular season only
#For me, a player's contribution is best measured by the amounts of points he scored, except for defenders and rucks.

#For this, I decided it was easier to make a function which would return the relevant stats for each position for each team, 
#sorting them from most time played to lowest

with_points = stats1.where('Round', are.containing('R')).with_column('Points', (6*regular_season_stats.column('Goals'))+regular_season_stats.column('Behinds'))

def contribution(team, position, year):
    #isolated to specific position, team, and year
    specific = with_points.where('Position', are.containing(position)).where('Team', team).where('Season', year)
    if position == 'Forward': #picking relevant stats for each position
        relevant_stats = specific.select('Player', 'Points', 'Disposals', 'Marks', 'Goals', 'Behinds', 'Inside50s', 'FreesFor', 'BrownlowVotes')
    if position == 'Midfield':
        relevant_stats = specific.select('Player', 'Points', 'Disposals', 'Marks', 'Goals', 'Behinds', 'Tackles', 'GoalAssists','BrownlowVotes')
    if position == 'Ruck':
        relevant_stats = specific.select('Player', 'Points', 'Disposals', 'Marks', 'Hitouts', 'Tackles', 'Clangers', 'OnePercenters', 'BrownlowVotes')
    if position == 'Defender':
        relevant_stats = specific.select('Player', 'Points', 'Tackles', 'Clangers', 'OnePercenters', 'BrownlowVotes')
    #averages out the percent of the game played per player
    percent_played = specific.select('Player', 'PercentPlayed').group('Player', np.average)
    #sums up other stats:
    total_position_stats = relevant_stats.group('Player', np.sum)
    #finds percent of points scored in a season for a player
    percent_score = 100*(total_position_stats.column('Points sum')/results.where('Team', team).where('Year', year).column('Pts Scored').item(0))
    #combines everything together
    question_three = percent_played.with_column('Percent of Team Score', percent_score).join('Player', total_position_stats)
    return question_three.sort('PercentPlayed average', descending = True) #sorts by Percent Played

#This shows that the biggest contributer at Forward for Adelaide in 2018 was Josh Jenkins
contribution('Adelaide', 'Forward', 2018)
```

```python
#4) Write code to help determine which players/teams had the most playoff success.
#For me, playoff success can be quantified in two ways, playoff wins and championships
#Let's start with championships
#Here, we are making a table that contains every championship team
champs = stats1.where('Round', are.equal_to('GF')).where('WinLoss', are.equal_to('W'))
champ_player_numbers = champs.group('Team')
champ_count = champ_player_numbers.with_column('Championships', champ_player_numbers.column('count')/22).drop('count')
champ_count #From this, we see that Hawthorn has dominated by winning three of the seven Grand Finals, nearly half!
```

```python
#We can do something similar to determine which PLAYERS have won the most championships
champ_players = champs.group('Player')
champ_players.sort('count', descending = True) #Here we see that all these players won three championships, meaning they are all
                                               #probably on Hawthorne
```

```python
#Next, let's see which teams have won the most playoff matches
playoff_wins = stats1.where('Round', are.containing('F')).where('WinLoss', 'W')
players_per_win = playoff_wins.group('Team')
playoff_wins_per_team = players_per_win.with_column('Playoff Wins', players_per_win.column('count')/22).drop('count')
playoff_wins_per_team.sort('Playoff Wins', descending = True) #Once again, we see that Hawthorne has been dominant
```

```python
#Once again, we can use a similar method to see which players have won the most playoff games
wins_per_player = playoff_wins.group('Player')
wins_per_player.sort('count', descending = True)
#Here we see that Sam Mitchell has won more playoff games than anyone else in AFL history
#The other table reveals that he has three championships as well
```

```python
#5) Develop your own assessment method of player performance by position and write code to rank players by position by year 
#   on that basis.

#For this question, I think the contributions() function I wrote for question three ranks players well based off of their team,
# so I decided to implement something similar.

def position_ranks(position, year):
    specific = with_points.where('Position', are.containing(position)).where('Season', year)
    if position == 'Forward': #picking relevant stats for each position
        relevant_stats = specific.select('Player', 'Points', 'Disposals', 'Marks', 'Goals', 'Behinds', 'Inside50s', 'FreesFor', 'BrownlowVotes')
    if position == 'Midfield':
        relevant_stats = specific.select('Player', 'Points', 'Disposals', 'Marks', 'Goals', 'Behinds', 'Tackles', 'GoalAssists','BrownlowVotes')
    if position == 'Ruck':
        relevant_stats = specific.select('Player', 'Points', 'Disposals', 'Marks', 'Hitouts', 'Tackles', 'Clangers', 'OnePercenters', 'BrownlowVotes')
    if position == 'Defender':
        relevant_stats = specific.select('Player', 'Points', 'Tackles', 'Clangers', 'OnePercenters', 'BrownlowVotes')
    percent_played = specific.select('Player', 'PercentPlayed').group('Player', np.average)
    total_position_stats = relevant_stats.group('Player', np.sum)
    question_five = percent_played.join('Player', total_position_stats).sort('BrownlowVotes sum', descending = True)
    return question_five #This time, I decided to rank the players by the number of Brownlow Medal votes they received
                         # as the award is a good indication of who was the best player.

position_ranks('Forward', 2018)
```
