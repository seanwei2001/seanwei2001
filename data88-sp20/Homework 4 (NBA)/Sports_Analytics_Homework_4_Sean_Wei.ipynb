```python
import numpy as np
import pandas as pd
import datascience as ds
```

```python
stats = pd.read_csv("Seasons_Stats.csv")
stats
```

```python
tot = stats.drop(stats.loc[stats["Tm"].isin(["TOT"])].index)
tot
```

```python
def is_hof(player):
    return (str(player)[-1]=="*")

hof = stats.loc[stats["Player"].map(is_hof)]
hof
```

```python
add_stats = hof[["Player", "ORB", "DRB", "TRB", "AST", "STL", "BLK", "TOV", "PF", "PTS"]].groupby("Player").apply(sum)
rate_stats = hof[['Player','FT%', 'FG%']].groupby('Player').apply(np.average)
add_stats
```
